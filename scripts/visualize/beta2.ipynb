```python
market_returns['Market'] = market
```

```python
market_returns['Market'].fillna(0, inplace=True)
```

```python
# stock_returns = df['StockReturns']
market_returns_col = market_returns['Market']
beta_dict = {}
for i in market_returns.columns:
    covariance = np.cov(market_returns[i], market_returns_col)[0, 1]
    variance_market = np.var(market_returns_col)
    beta = covariance / variance_market
    beta_dict[i] = beta
```

```python
df_beta = pd.DataFrame.from_dict(beta_dict, orient='index', columns=['Beta']).sort_values(by='Beta', ascending=True)
df_beta= df_beta[:-1]
# Calculate average beta
# average_beta = df_beta['Beta'].mean()
average_beta = 1

# Plot the beta values and average line
plt.figure(figsize=(20, 10))
plt.bar(df_beta.index, df_beta['Beta'], color='blue', label='Beta')
plt.axhline(y=average_beta, color='red', linestyle='--', label='Average Beta')

plt.title('Beta of Cryptocurrencies')
plt.xlabel('Cryptocurrency')
plt.xticks(rotation=45)
plt.ylabel('Beta')
plt.grid(True)
# plt.legend()
plt.show()
```

```python
# split into boom and bust periods
boom1 = market_returns.loc['2020-03-16':'2021-05-11']
boom2 = market_returns.loc['2021-07-20':'2021-11-09']
boom3 = market_returns.loc['2022-06-30':'2022-08-12']
boom4 = market_returns.loc['2022-11-21':'2023-01-23']
bust1 = market_returns.loc['2021-05-11':'2021-07-20']
bust2 = market_returns.loc['2021-11-09':'2022-06-30']
bust3 = market_returns.loc['2022-08-12':'2022-11-21']
```

```python
def get_beta_dict(market_returns):
    market_returns_col = market_returns['Market']
    beta_dict = {}
    for i in market_returns.columns:
        covariance = np.cov(market_returns[i], market_returns_col)[0, 1]
        variance_market = np.var(market_returns_col)
        beta = covariance / variance_market
        beta_dict[i] = beta
    return beta_dict
```

```python
boom1_beta = get_beta_dict(boom1)
boom2_beta = get_beta_dict(boom2)
boom3_beta = get_beta_dict(boom3)
boom4_beta = get_beta_dict(boom4)
bust1_beta = get_beta_dict(bust1)
bust2_beta = get_beta_dict(bust2)
bust3_beta = get_beta_dict(bust3)
```

```python
# create a dataframe of the boom1_beta and boom2_beta dictionaries
boom1_df = pd.DataFrame.from_dict(boom1_beta, orient='index', columns=['Beta']).sort_values(by='Beta', ascending=True)
boom2_df = pd.DataFrame.from_dict(boom2_beta, orient='index', columns=['Beta']).sort_values(by='Beta', ascending=True)
boom3_df = pd.DataFrame.from_dict(boom3_beta, orient='index', columns=['Beta']).sort_values(by='Beta', ascending=True)
boom4_df = pd.DataFrame.from_dict(boom4_beta, orient='index', columns=['Beta']).sort_values(by='Beta', ascending=True)
# merge these dataframes into 1 daatframe
boom_df = pd.concat([boom1_df, boom2_df, boom3_df, boom4_df], axis=1)
boom_df['mean'] = boom_df.mean(axis=1)
```

```python
# add an average line to the plot
pd.DataFrame(boom_df['mean'][:-1]).sort_values(by='mean', ascending=False).plot(kind='bar', figsize=(15, 6), grid=True, legend = False, title = 'Average Beta of Cryptocurrencies for Boom Periods')
# plt.axhline(y=1, color='r', linestyle='-')
```
