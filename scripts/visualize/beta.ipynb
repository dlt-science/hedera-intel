```python
# beta calculation functions
def covariance(self, SPY_hist, hist2):

    #SPY_closes = SPY_hist['Close'].to_list()
    SPY_closes = SPY_hist
    length = len(SPY_closes)

    SPY_average = 0
    for p in SPY_closes:
        SPY_average += p
    SPY_average = SPY_average / length

    #hist2_closes = hist2['Close'].to_list()
    hist2_closes = hist2
    length = len(hist2_closes)

    hist2_average = 0
    for day in hist2_closes:
        hist2_average += day
    hist2_average = hist2_average / length

    total = 0
    for i in range(len(SPY_hist)):
        total += ((SPY_closes[i] - SPY_average) * (hist2_closes[i] - hist2_average))

    return total / (length - 1)

def variance(self, SPY_hist):

    #SPY_closes = SPY_hist['Close'].to_list()
    SPY_closes = SPY_hist
    length = len(SPY_closes)

    SPY_average = 0
    for p in SPY_closes:
        SPY_average += p
    SPY_average = SPY_average / length

    sum = 0
    for i in SPY_closes:
        dif = i - SPY_average
        sum += (dif ** 2)

    return (sum / (length-1))


def beta(self, SPY_hist, hist2):

    _covariance = self.covariance(SPY_hist=SPY_hist, hist2=hist2)

    _variance = self.variance(SPY_hist=SPY_hist)

    return (_covariance / _variance)

def percent_change(self, hist):

    closes = hist['Close'].to_list()
    percent_change = []
    for i in range(len(closes)):
        if i != 0:
            percent_change.append(((closes[i-1] - closes[i]) / closes[i-1]) * 100)

    return percent_change

def get_beta(self, securities):

    SPY_hist = yf.Ticker('SPY').history(period='5y', interval='1mo')
    SPY_pc = self.percent_change(hist=SPY_hist)
    betas = {}

    for security in securities:
        security_hist = yf.Ticker(security).history(period='5y', interval='1mo')
        security_pc = self.percent_change(hist=security_hist)
        b = self.beta(SPY_hist=SPY_pc, hist2=security_pc)
        betas[security] = b

    return betas
```

```python
# function to compute dtw matrix
def compute_dtw_matrix(dataframes):
    num_dataframes = len(dataframes)
    dtw_matrix = np.zeros((num_dataframes, num_dataframes))

    for i in range(num_dataframes):
        for j in range(i, num_dataframes):
            distance = dtw(dataframes[i]['value'], dataframes[j]['value'])
            dtw_matrix[i, j] = distance
            dtw_matrix[j, i] = distance  # Since DTW is symmetric

    return dtw_matrix
```

```python
columns_with_nan = merged_prices.columns[merged_prices.isna().any()].tolist()

# Drop columns with NaN values
df_dropped = merged_prices.dropna(axis=1)

# Print the list of dropped columns
print("Dropped columns:", columns_with_nan)

# Display the DataFrame after dropping columns
print("DataFrame after dropping columns:")
print(df_dropped)
```

```python
# convert the values in merged prices to get market return
market_returns = df_dropped.pct_change()
```

```python
# create a column in market returns that is the average of all the market returns
market_returns['Market'] = market_returns.mean(axis=1)
```

```python
# remove the first row from the dataframe
market_returns = market_returns.iloc[1:]
```

```python
# stock_returns = df['StockReturns']
market_returns_col = market_returns['Market']
beta_dict = {}
for i in market_returns.columns:
    covariance = np.cov(market_returns[i], market_returns_col)[0, 1]
    variance_market = np.var(market_returns_col)
    beta = covariance / variance_market
    beta_dict[i] = beta
```

```python
df_beta = pd.DataFrame.from_dict(beta_dict, orient='index', columns=['Beta']).sort_values(by='Beta', ascending=False)

# Calculate average beta
# average_beta = df_beta['Beta'].mean()
average_beta = 1

# Plot the beta values and average line
plt.figure(figsize=(20, 10))
plt.bar(df_beta.index, df_beta['Beta'], color='blue', label='Beta')
plt.axhline(y=average_beta, color='red', linestyle='--', label='Average Beta')

plt.title('Beta of Cryptocurrencies')
plt.xlabel('Cryptocurrency')
plt.xticks(rotation=45)
plt.ylabel('Beta')
plt.grid(True)
# plt.legend()
plt.show()
```
